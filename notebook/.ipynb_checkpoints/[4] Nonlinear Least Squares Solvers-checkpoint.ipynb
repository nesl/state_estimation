# Tutorial #4 :: Nonlinear Least Squares Solvers ( in prog.)

The previous estimation techniques were very powerful solutions to real-time estimation requirements.  In some scenarios, we have a lot of data (observations) and we want to make an informed estimate taking into consideration *all* of those observations.  For example, In our particle filter scenario we were out of luck if we encountered a large time period where we didn't have any range estimates from our anchors. When we *did* receive a new range measurement, the position estimate would jump to the new estimate in a discontinuous fashion.  If we do not need a causal estimation technique, we can use future observations to constrain earlier estimates as well--we could, e.g., interpolate between guesses.

This is where least squares solvers come in handy, including nonlinear least squares solvers in particular. But least squares solvers are simple and boring, right? We use them to draw a straight line through some points, or to fit a nice exponential curve through a set of data. Yes, we do, but we can do *so* much more with them as well.

In this section, we won't deal with any of the math behind least squares solvers, treating them instead as a black box. The key point, however, will be that we want to estimate a state vector $x$ such that it minimizes the sum of squared errors. The errors themselves are dictated by functions of the state vector that we provide as developers.

To begin, we're going to do a really naive estimation technique based on our innertial data--we're going to detect steps and claim that these steps indicate a movement in the forward direciton of some fixed stride length. Then we'll update our yaw angle (our rotation about the vertical axis) by looking at our vertical gyroscope axis.

First thing's first: let's load some data.

```python
from scipy import signal
import numpy as np

# Load inertial data
data_imu = np.array(np.loadtxt('../data/imudata.txt', skiprows=1))
tstamps = data_imu[:,2]
acc_xyz = data_imu[:,3:6]
gyr_xyz = data_imu[:,6:9]

# Filter accel. to get gravity vector and linear accel.
b, a = signal.butter(N=4, Wn=4.0/(100/2), btype='lowpass', analog=False, output='ba')
grav_xyz = signal.filtfilt(b,a,acc_xyz, axis=0)
lin_xyz = acc_xyz - grav_xyz
```

Next, we'll detect steps like we did for the ZUPT correction mechanism in the EKF tutorial.

```python
import matplotlib
import matplotlib.pyplot as plt

# First, some Jupyter plotting config
%matplotlib inline
%config InlineBackend.figure_formats = {'svg',}

# get total energy in acceleration signal
lin_energy = np.linalg.norm(lin_xyz, axis=1)

# filter these energy spikes so we can easily detect peaks
b, a = signal.butter(N=10, Wn=4.0/(100/2), btype='lowpass', analog=False, output='ba')
lin_energy_filt = signal.filtfilt(b,a,lin_energy, axis=0)

# peak detection
peaks = np.array(signal.find_peaks_cwt(lin_energy_filt, np.arange(1,10), min_snr=1))
peakvals = lin_energy_filt[peaks]
footstep_idxs = peaks[peakvals > 4.0]

plt.plot(lin_energy_filt, 'r', markersize=3)   
plt.plot(footstep_idxs, lin_energy_filt[footstep_idxs], 'bo', markersize=5)
plt.title('Position estimate via EKF')
plt.xlabel('Index')
plt.ylabel('Acceleration Norm')
plt.legend(['Energy', 'Footstep'])
plt.show()
```

Next we want to estimate our rotation about the vertical axis based on our gyroscope signal. We'll have to perform tilt correction like before, in case our sensor (foot) isn't perfectly horizontal.

```python
gyr_z_tilt = np.array([]).reshape(0,1)

for idx in range(acc_xyz.shape[0]):
    # grav vectors
    grav = grav_xyz[idx,:]
    gx, gy, gz = grav
    # gyro vector
    gyro = gyr_xyz[idx,:].reshape(3,1)
    
    # get roll, pitch, & yaw
    pitch = np.arcsin( gx/np.linalg.norm(grav) )
    roll = np.arcsin( -gy/np.linalg.norm(grav) )
    
    Rx = np.matrix([
            [1, 0, 0],
            [0, np.cos(roll), -np.sin(roll)],
            [0, np.sin(roll), np.cos(roll)],
        ])
    Ry = np.matrix([
            [np.cos(pitch), 0, np.sin(pitch)],
            [0, 1, 0],
            [-np.sin(pitch), 0, np.cos(pitch)],
        ])

    # yaw-pitch-roll form
    R_b2g = Ry*Rx
    gyr_xyz_tilt = R_b2g*gyro
            
    # append to array
    gyr_z_tilt = np.vstack([gyr_z_tilt, gyr_xyz_tilt.item(2)])
    
plt.plot(-(1.0/100)*np.cumsum(gyr_z_tilt), 'b')   
plt.plot(-(1.0/100)*np.cumsum(gyr_xyz[:,2]), 'r')   
plt.title('Estimating Yaw Angles')
plt.xlabel('Index')
plt.ylabel('Yaw Angle (rad)')
plt.legend(['Tilt compensated', 'Using raw gyro'])
plt.show()
```

Cool, so now we have a nice estimate of our yaw rotation and our footsteps. We'll now fuse the two so we can have an initial estimate of our path shape. We'll have to make an initial asssumption for both the position and orientation at the start.

```python
# Our initial path starts at the origin
path_xy = np.array([0,0]).reshape(1,2)
path_yaw = np.array([-np.pi/2]).reshape(1,1)
yaw_scalar = 1.2

# Stride length
slength = 0.50

t_last = tstamps[0]

for idx in range(1,len(tstamps)):
    # time difference
    dt = tstamps[idx] - t_last
    t_last = tstamps[idx]
    
    # integrate gyro to get angle
    yaw_new = path_yaw[idx-1] + -dt*gyr_xyz[idx,0]*yaw_scalar
    path_yaw = np.vstack([path_yaw, yaw_new])
    
    # if a step is detected, advance the path by stride length
    # given current orientation
    if idx in footstep_idxs:
        dxy = np.array([np.cos(yaw_new)*slength, np.sin(yaw_new)*slength]).reshape(1,2)
        new_xy = path_xy[-1] + dxy
        path_xy = np.vstack([path_xy, new_xy])
        
plt.plot(path_xy[:,0], path_xy[:,1], 'bo:', markersize=3)  
plt.plot(path_xy[0,0], path_xy[0,1], 'gs', markersize=7)
plt.title('Rough Position Estimate')
plt.xlabel('X Position(m)')
plt.ylabel('Y Position(m)')
plt.legend(['Path', 'Start'])
plt.show()
```

This is pretty bad, but this is where the least squares solver comes in to help us out. Let's say we have a couple of extra observations that we can use to help constrain the system. In fact, let's say we have precisely two new observations--the starting position, and the ending position. 

Note a couple of little hacks we had to do to get a decent shape in the code above--we have some scalar, because the gyroscope doesn't translate perfectly into a nice angular estimate, and we're not really sure about stride-length. The goal will be to use a nonlinear least squares solver to find a stridelength and a gyroscope scalar such that our estimate is much more accurate given new observations.
